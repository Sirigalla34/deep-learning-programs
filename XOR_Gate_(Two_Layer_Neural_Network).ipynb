```python
import numpy as np

x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
t = np.array([0,1,1,0])

m = len(t)

max_iter = 9000
learn_rate = 0.3

w1 = np.random.uniform(-1,1)*0.2
w2 = np.random.uniform(-1,1)*0.2
w3 = np.random.uniform(-1,1)*0.2
w4 = np.random.uniform(-1,1)*0.2
w5 = np.random.uniform(-1,1)*0.2
w6 = np.random.uniform(-1,1)*0.2

b1 = 0
b2 = 0
b3 = 0

losses = []

def sigmoid(z):
    return 1/(1+np.exp(-z))

def sigmoid_derivative(s):
    return s*(1-s)

for _ in range(max_iter):

    dw1=dw2=dw3=dw4=dw5=dw6=0
    db1=db2=db3=0
    mse = 0

    for i in range(m):

        zh1 = x1[i]*w1 + x2[i]*w2 + b1
        zh2 = x1[i]*w3 + x2[i]*w4 + b2

        h1 = sigmoid(zh1)
        h2 = sigmoid(zh2)

        zout = h1*w5 + h2*w6 + b3
        y = sigmoid(zout)

        error = t[i] - y
        mse += error**2

        d_out = error * sigmoid_derivative(y)

        dw5 += d_out * h1
        dw6 += d_out * h2
        db3 += d_out

        d_h1 = d_out * w5 * sigmoid_derivative(h1)
        d_h2 = d_out * w6 * sigmoid_derivative(h2)

        dw1 += d_h1 * x1[i]
        dw2 += d_h1 * x2[i]
        db1 += d_h1

        dw3 += d_h2 * x1[i]
        dw4 += d_h2 * x2[i]
        db2 += d_h2

    w1 += learn_rate*(dw1/m)
    w2 += learn_rate*(dw2/m)
    w3 += learn_rate*(dw3/m)
    w4 += learn_rate*(dw4/m)
    w5 += learn_rate*(dw5/m)
    w6 += learn_rate*(dw6/m)

    b1 += learn_rate*(db1/m)
    b2 += learn_rate*(db2/m)
    b3 += learn_rate*(db3/m)

    losses.append(mse/m)

print("Final Weights:")
print(w1,w2,w3,w4,w5,w6)
print("Bias:",b1,b2,b3)

print("\nPredictions:")
for i in range(m):

    h1 = sigmoid(x1[i]*w1 + x2[i]*w2 + b1)
    h2 = sigmoid(x1[i]*w3 + x2[i]*w4 + b2)

    y = sigmoid(h1*w5 + h2*w6 + b3)

    print(x1[i],x2[i],"->",round(y))
```
