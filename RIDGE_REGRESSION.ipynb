```python
import numpy as np
X = np.array([[1], [2], [3], [4]])
y = np.array([2, 4, 6, 8])
lr = 0.01
epochs = 1000
lam = 0.1
w = 0.0
for _ in range(epochs):
    y_pred = w * X.flatten()
    error = y_pred - y
    dw = (2 * np.mean(error * X.flatten())) + (2 * lam * w)
    w = w - lr * dw

print("Ridge Weight:", w)
```
