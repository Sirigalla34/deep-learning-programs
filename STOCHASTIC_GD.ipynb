```python
import numpy as np
X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

lr = 0.01
epochs = 50
w = 0.0

for _ in range(epochs):
    for i in range(len(X)):
        y_pred = w * X[i]
        error = y_pred - y[i]
        w = w - lr * (2 * error * X[i])

print("SGD Weight:", w)
```
