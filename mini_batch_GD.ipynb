```python
import numpy as np

X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

lr = 0.01
epochs = 100
batch_size = 2
w = 0.0

for _ in range(epochs):
    for i in range(0, len(X), batch_size):
        Xb = X[i:i+batch_size]
        yb = y[i:i+batch_size]

        y_pred = w * Xb
        error = y_pred - yb
        dw = 2 * np.mean(error * Xb)
        w = w - lr * dw

print("Mini-Batch GD Weight:", w)
```
