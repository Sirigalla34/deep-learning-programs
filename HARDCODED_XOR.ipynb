```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

print("XOR GATE - 2 LAYER PERCEPTRON (HARD CODED)")
W1 = np.array([[1, 1],
               [1, 1]])
b1 = np.array([[-0.5, -1.5]])
W2 = np.array([[1],             [-2]])
b2 = np.array([[-0.5]])
def step(x):
    return (x > 0).astype(int)
def forward(X):
    z1 = np.dot(X, W1) + b1
    a1 = step(z1)
    z2 = np.dot(a1, W2) + b2
    y_hat = step(z2)
    return z1, a1, z2, y_hat

z1, a1, z2, preds = forward(X)

print("\nPREDICTIONS")
for i in range(len(X)):
    print(f"{X[i]} → Hidden: {a1[i]} → Output: {preds[i][0]} | Actual: {y[i][0]}")
```
