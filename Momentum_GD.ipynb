```python
import numpy as np

X = np.array([1, 2, 3, 4])
y = np.array([2, 4, 6, 8])

lr = 0.01
epochs = 100
beta = 0.9
w = 0.0
v = 0.0

for _ in range(epochs):
    y_pred = w * X
    error = y_pred - y
    dw = 2 * np.mean(error * X)
    v = beta * v + lr * dw
    w = w - v

print("Momentum GD Weight:", w)
```
