```python
import numpy as np

x1 = np.array([0,0,1,1])
x2 = np.array([0,1,0,1])
t = np.array([0,0,0,1])

m = len(t)

max_iter = 4000
learn_rate = 0.1

w1 = np.random.uniform(-1,1)*0.2
w2 = np.random.uniform(-1,1)*0.2
b = 0

def sigmoid(z):
    return 1/(1+np.exp(-z))

def sigmoid_derivative(s):
    return s*(1-s)

for _ in range(max_iter):

    dw1 = 0
    dw2 = 0
    db = 0

    for i in range(m):

        z = x1[i]*w1 + x2[i]*w2 + b
        y = sigmoid(z)

        error = t[i] - y
        d = error * sigmoid_derivative(y)

        dw1 += d * x1[i]
        dw2 += d * x2[i]
        db += d

    w1 += learn_rate * (dw1/m)
    w2 += learn_rate * (dw2/m)
    b += learn_rate * (db/m)

print("Final Weights and Bias:")
print("w1 =",w1)
print("w2 =",w2)
print("b =",b)

print("\nPredictions:")
for i in range(m):
    z = x1[i]*w1 + x2[i]*w2 + b
    y = sigmoid(z)
    print(x1[i],x2[i],"->",round(y))
```
