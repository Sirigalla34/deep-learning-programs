```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])
y = np.array([[0],[1],[1],[1]])
print("OR Gate Single Perceptron")
weights = np.array([[1],[1]])
bias = -0.5
w1 = weights[0][0]
w2 = weights[1][0]
print("Weight w1 =", w1)
print("Weight w2 =", w2)
print("Bias b =", bias)
print("Equation : {}*x1 + {}*x2 + ({})".format(w1,w2,bias))
def step_function(x):
    return 1 if x > 0 else 0
def perceptron_predict(x, weights, bias):
    linear = np.dot(x, weights) + bias
    return (linear > 0).astype(int)
print("\nPredictions")
linear_values = np.dot(X, weights) + bias
predictions = perceptron_predict(X, weights, bias)
for i in range(len(X)):
    linear_val = linear_values[i][0]
    pred = predictions[i][0]
    actual = y[i][0]
    print(f"Input {X[i]} -> Linear:{linear_val:.2f}  Pred:{pred}  Actual:{actual}")
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
plt.scatter(X[y.flatten()==0][:,0],
            X[y.flatten()==0][:,1],
            s=600, marker='o',
            edgecolor='black',
            linewidth=2,
            label='Class 0',
            color='red')
plt.scatter(X[y.flatten()==1][:,0],
            X[y.flatten()==1][:,1],
            s=600, marker='s',
            edgecolor='black',
            linewidth=2,
            label='Class 1',
            color='blue')

x_range = np.linspace(-0.3,1.3,200)

x2_boundary = (-w1*x_range - bias)/w2

plt.plot(x_range, x2_boundary, 'k--', label='Decision Boundary')

plt.fill_between(x_range,-0.3,x2_boundary,alpha=0.15,color='red')
plt.fill_between(x_range,x2_boundary,1.3,alpha=0.15,color='blue')

plt.xlim(-0.3,1.3)
plt.ylim(-0.3,1.3)

plt.xlabel("Input 1 (x1)")
plt.ylabel("Input 2 (x2)")
plt.title("OR Gate Decision Boundary")

plt.grid(True, alpha=0.3)
plt.legend()
plt.subplot(1,2,2)

samples = ['[0,0]','[0,1]','[1,0]','[1,1]']
linear_vals = linear_values.flatten()

colors = ['red' if y[i][0]==0 else 'blue' for i in range(len(y))]

plt.bar(samples, linear_vals,
        color=colors,
        alpha=0.8,
        edgecolor='black',
        linewidth=1)

plt.axhline(y=0,
            color='black',
            linestyle='--',
            linewidth=1,
            label='Threshold')

plt.title("Linear Output Values")
plt.ylabel("Value")

plt.legend()

plt.show()
```
